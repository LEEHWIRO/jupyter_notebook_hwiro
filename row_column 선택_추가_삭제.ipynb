```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris[1:4]
```

```python
# loc() : 레이블(인덱스의 이름 혹은 컬럼명)로 행과 열을 선택하고자 할 때는 loc[]속성을 사용한다.

iris.loc[1:3]
```

```python
iris.loc[[1,2], 'sepal length (cm)': 'petal length (cm)']
```

```python
score = pd.DataFrame({'국어' : [100, 80], '수학' : [75, 90], '영어' : [90, 95]}, index=['장화', '홍련'])
score.loc['홍련', '영어'] = 100
score['국어'] = score['국어'] - 5
score
```

```python
# append() : 데이터프레임에 한 행을 추가 인수 : DataFrame, Series, Dictionary

new_students = pd.DataFrame({'국어':[70,85], '수학':[65,100], '영어':[95,65]}, index=['콩쥐','팥쥐'])
score = score.append(new_students)
new_student1 = pd.Series({'국어':85, '수학':55, '영어':95}, name='해님')
score = score.append(new_student1)
score
```

```python
# ignore_index

new_student2 = [{'국어':75, '수학':80, '영어':80}]
score.append(new_student2, ignore_index=True)
```

```python
# column 추가

science = [80,70, 90, 85, 75]
score['과학'] = science
score['학년'] = 1
score
```

```python
score['과학'] = score['과학'] + 5
score['총점'] = score['국어'] + score['수학'] + score['영어'] + score['과학']
score
```

```python
# DataFrame.drop(index=None, columns=None, inplace=False)
# index : 삭제할 행의 인덱스 혹은 인덱스의 리스트를 지정
# columns : 삭제할 컬럼의 이름 혹은 컬럼명의 리스트를 지정
# inplace : False일 경우 작업 수행의 결과를 복사본으로 반환하고, True이면 작업을 수행하고 None을 반환

score.drop('장화', inplace=True)
score.drop(columns = ['과학', '학년', '총점'], inplace=True)
score
```
